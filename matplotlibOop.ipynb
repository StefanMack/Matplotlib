### Unterscheidung matplotlib.pyplot und matpltolib.pylab
Mit pyplot und pylab können grundsätzlich die selben Grafiken erstellt werden. Es handelt sich hierbei nur um zwei unterschiedliche Schnittstellen ("Interfaces") von matplotlib. pylab ist speziell als MATLAB-Ersatz konzipiert. D.h. hier wird eine fast identsiche Syntax wie bei MATLAB für die Datenvisualisierung verwendet. Mit den so erzeugten Grafiken kann man anschließend mit der Maus interaktiv agieren (z.B. hinein zoomen oder Werte ablesen). pylab ist wie MATLAB *nicht* streng objektorientiert.  
**pyplot ist der "pythonic way"**: In pyplot hat man sich von er MATLAB-Syntax gelöst und pyplot *kann* -  muss aber nicht - konsistent objektorientiert verwendet werden: Eine Darstellung ist ein Objekt, welches z.B. Plots als Objekte enthält, die wiederum z.B. Achsen als Objekte enthalten. Bei der Erstellung eines Grafs werden somit die entsprechenden Objekte instanziert.  
  
**In diesem Jupyter Notebook wird ausschließlich ``matplotlib.pyplot`` verwendet.**  

Innerhalb ``matplotlib.pyplot`` kann sowohl recht *MATLAB-ähnlich prozedural* als auch *konsequent oibjektorientiert* gecodet werden. Gerade für Anfänger ist dies leider verwirrend.  In ersteren Fall gibt es immer eine aktive Darstellung, auf die die Befehle wirken. Bei der objektorientierten Verwendung, werden den verschiedenen Darstellungen, Plots usw. jeweils Variablen zugeordnet, die auf diese Instanzen verweisen. **In diesem Notebook wird ausschließlich der objektorientierte Zugriff über Instanzvariablen verwendet.**

### Einfacher x-y-Linienplot, Erzeugen einer Plotgrafik
Üblicherweise werden die Variablennamen ``fig`` für die Darstellung und ``ax`` für den Plot verwendet. Davon sollte man im Sinne der Codelesbarkeit nicht abweichen.  
``linestyle`` spezifiziert Linieart: ``'-'``, ``'--'``, ``'-.'``, ``':'``
``color`` spezifiziert Farbe: ``'blue'``, ``'r'``, ``'#FFDD44'``, ``(1.0,0.2,0.3)``, ``'C4'``, ``'0.75'`` wobei die Farbe über deren ausgeschriebenen Namen, über den entsprechenden Buchstaben, über einen dreidimensionalen RGB-Hexvektor bzw. RGB-FloatVelktor, über eine in der Formatvorlage vorgegebene Farbreihenfolge oder über einen Grauwert quantifiziert wird.  

Möchte man eine Darstellung als Grafik abspeichern, so verwendet man dafür die Variable ``fig``.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig, ax = plt.subplots() # Darstellung (fig) und Plot (ax) instanzieren
ax.plot(x,y,linestyle=':',color='C4', label='Sinusfunktion');
ax.set_xlabel('x-Wert') # Beschriftung x-Achse
ax.set_ylabel('y-Wert')
ax.set_title('Sinusplot')
ax.set_xlim(0,9) # Bereich x-Achse
ax.set_ylim(-0.5,1)
ax.set_yticks(np.arange(-1, 1, step=0.25)) # Teilung der y-Achse
ax.grid() # Gitter erzeugen
ax.legend() # Legende mit Werten aus ax.plot(label='')
fig.savefig('my_figure.png') # Abpeichern einer png-Grafik der gesamten Darstellung
# Semikolon, damit keine zusätzliche Textausgabe "[<matplotlib.lines.Line2D..."
# plt.show() # nur außerhalb Jupyter Notebook nötig
```

Alle ``ax.set_...`` Befehle können übrigens in einen Befehl ``ax.set(xlim=(0, 9), ...)`` zusammengefasst werden.  
Als Dateiformate kommen folgende in Frage: 'eps', 'jpeg', 'jpg', 'pdf', 'pgf', 'png', 'ps', 'raw', 'rgba', 'svg', 'svgz', 'tif', 'tiff'. Sie können über den Befehl ``fig.canvas.get_supported_filetypes()`` abgerufen werden.

### Einfacher xy-Punkteplot  
Punkteplots werden analog zu den Linienplots erstellt. Für die Symbolart stehen folgende Arten zur Verfügung:  
``'+'``, ``'o'``, ``'.'``, ``','``, ``'x'``, ``'+'``, ``'v'``, ``'^'``, ``'<'``, ``'>'``, ``'s'``, ``'d'``  
Punkte und Linien können auch kombiniert werden. Beispiel: ``':o'`` für eine gestrichelte Linie zwischen den Punkten als Vollkreise.

```python
x = np.linspace(0, 10, 20)
y = np.sin(x)
fig, ax = plt.subplots() # Darstellung (fig) und Plot (ax) instanzieren
ax.plot(x,y,':o',color='C0', label='Sinusfunktion'); # gestrichelte Linie und runde Symbolpunkte
ax.set_xlabel('x-Wert') # Beschriftung x-Achse
ax.set_ylabel('y-Wert')
ax.set_title('Sinusplot mit Punkten')
ax.grid(linestyle=':') # Gitter erzeugen
ax.legend() # Legende mit Werten aus ax.plot(label='')
```

### Mehrere x-y-Plots im selben Diagramm
Hierbei wird für jede Spur ein eigener ``ax.plot()``-Befehl verwendet. Möchte man die verschiedenen Spuren nicht explizit formatieren, so kann auch ein gemeinsamer Plotbefehl ``ax.plot(x, y0, x, y1, x ,y2)`` verwendet werden.


```python
x = np.linspace(0, 10, 100)
y0 = np.sin(x)
y1 = np.sin(1.1*x)
y2 = np.sin(1.2*x)

fig, ax = plt.subplots()
ax.plot(x,y0,linestyle=':',color='C0', label='Sinusfunktion 0');
ax.plot(x,y1,linestyle='-',color='C1', label='Sinusfunktion 1');
ax.plot(x,y2,linestyle='--',color='C2', label='Sinusfunktion 2');
ax.set_xlabel('x-Wert') # Beschriftung x-Achse
ax.set_ylabel('y-Wert')
ax.set_title('Drei Sinusplots')
ax.grid(color='b', linestyle=':', linewidth=1) # Gitter erzeugen
ax.legend(); # Legende mit Werten aus ax.plot(label='')
```

### Mehrere Diagramme in einer Darstellung
Werden innerhalb einer Darstellung ``fig`` meherer Plots erzeugt, so wird die Variable ``ax`` zu einem NumPy-Array, dessen einzelnen Elemente den Zugriff auf den jeweiligen Plot ermöglichen.

```python
x = np.linspace(0, 10, 100)
y0 = np.sin(x)
y1 = np.sin(1.0*x)
y2 = np.sin(1.1*x)
y3 = np.sin(1.2*x)
y4 = np.sin(1.3*x)
y5 = np.sin(1.4*x)

fig, ax = plt.subplots(nrows=2, ncols=3, sharex='none', sharey='row', figsize=(8,4),constrained_layout=True)
# constrained_layout() damit keine überlappende Anordnung der einzelnen Plots, Beschriftungen usw.

ax[0,0].plot(x,y0,label='Sinus 0')
ax[0,1].plot(x,y1,label='Sinus 1')
ax[0,2].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)
ax[1,2].plot(x,y5)

fig.suptitle('Sechs Sinusplots', fontsize=16)

ax[0,0].legend()
ax[0,1].legend()
ax[0,0].set(xlabel='x-Werte', ylabel='y-Werte', title='Wichtig!!');
ax[1,0].set(xlabel='mehr x-Werte', ylabel='mehr y-Werte', title='Unwichtig');

#plt.tight_layout() # alternativ zu constrained_layout()
```

### Histogrammplots
Statt ``ax.plot()`` wird der Befehl ``ax.hist()`` verwendet, um Histogramme zu erzeugen.

```python
Daten0 = np.random.normal(0, 0.7, 1000)
Daten1 = np.random.normal(-1, 1.2, 1000)
Daten2 = np.random.normal(2, 2, 1000)

fig, ax = plt.subplots()
# 'bins' sind die Anzahl der "Töpfe", 'alpha' gibt die Transparenz an
ax.hist(Daten0, bins=30, alpha=0.4, histtype='stepfilled', color='C0',edgecolor='none', label='Referenz');
ax.hist(Daten1, bins=30, alpha=0.4, histtype='stepfilled', color='C1',edgecolor='none');
ax.hist(Daten2, bins=30, alpha=0.4, histtype='stepfilled', color='C2',edgecolor='none');
ax.set(xlabel='Temperatur ( °C)', ylabel='Häufigkeit', title='Temperaturmessung');
ax.legend();
```

### Fehlerbalken
Über die Methode ``errorbar()`` werden sowohl die Datenpunkte als auch die Fehlerbalken erzeugt. Mit ``fmt`` und ``color`` werden die Symbole bzw. die Farbe für die Datenpunkte festgelegt. ``capsize`` gibt die länge der Begrenzungsstriche der Fehlerbalken an.

```python
x = np.linspace(0, 10, 50)
dy = 0.6
y = 0.5*x + dy * np.random.randn(50)

fig, ax = plt.subplots()

ax.errorbar(x, y, yerr=dy, fmt='.', color='b', ecolor='k', elinewidth=1, capsize=2);
ax.grid(linestyle=':')
ax.set(xlabel='Zeit (h)', ylabel='Temperatur (°C)', title='Temperaturmessung');
```

### Plots von 3D-Daten
Sogenannte skalare Felder können innerhalb eines 2D-Plots über Farben oder Konturlinien dargestellt werden.  
Das skalare Feld ist über folgende Funktion definiert:

```python
def skal_feld(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = skal_feld(X, Y)
```

#### Konturplot
Analog zu den vorangegangenen Plotbeispielen wird der 3D-Plot über eine spezielle Methode von des Plotobjekts erzeugt: ``ax.contour()``

```python
fig, ax = plt.subplots(constrained_layout=True, figsize=(10, 6))
cs = ax.contour(X, Y, Z, colors='black');
ax.clabel(cs, inline=1, fontsize=8);
```

#### Falschfarbenplot
Entsprechen wird der 3D-Plot über die Methode ``ax.contourf()`` als Falschfarbenplot dargestellt. Für die Zuordnung der z-Werte zu den Farben wird ein "colorbar" als Legende eingefügt.  
Es gibt eine Vielzahl von "Colormaps" (``cmap``), die unter matplotlib.org zu finden sind.

```python
fig, ax = plt.subplots()
cs = ax.contourf(X, Y, Z, 20, cmap='jet'); 
fig.colorbar(cs);
```

### Gestaltung von Plots mittels Formatvorlagen
Damit nicht jede Ploteigenschaft explizit angegeben werden muss, kann man Formatvorlagen verwenden, welche bei matplotlib "Stylesheets" genannt werden. Welche hierfür zur Auswahl stehen, egibt der folgende Befehl aus:

```python
IPython_default = plt.rcParams.copy() # aktuell verwendetes Stylesheet sichern
plt.style.available[:6] # Ausgabe der ersten sechs verfügbaren Stylesheets
```

Nachfolgend wird ein Funktion erstellt, welche ein Histogramm und einen xy-Plot erstellt. Diese Funktion wird anschließend für unterschiedliche Stylesheets ausgeführt, um diese verlgeichend darzustellen.

```python
def hist_and_lines(): # Funktion zur Erstellung eines Histgramms und eines xy-Plots. Quelle: Python Data Sciende Handbook
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(13, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

```python
# reset rcParams falls vorher ein anderes Stylesheet geladen war
plt.rcParams.update(IPython_default);
hist_and_lines()
```

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```

```python
with plt.style.context('seaborn'):
    hist_and_lines()
```

Eine Ausnahme bezüglich der Syntax stellt die Formatvorlagen **xkcd** dar. Diese wird wie folgt verwendet:

```python
# Aktiviere XKCD - Modus
# plt.xkcd() # geht auch statt with plt.xkcd()
with plt.xkcd():
    hist_and_lines()
```
